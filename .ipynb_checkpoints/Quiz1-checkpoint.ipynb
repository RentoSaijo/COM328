1. Question: (a) For the ISLP (Wage) dataset what years are recorded here? (b) How many rows of data are recorded for each year on record? (c) Find the mean, min and max wage for each of these years on record. (d) Does the result make sense?

```python
# Load libraries.
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from ISLP import load_data

# Load Wage data.
Wage = load_data('Wage')

# (a) Find recorded years.
years = np.sort(Wage['year'].unique())
print('Years recorded:', years)

# (b) Find rows per year.
rows_per_year = Wage['year'].value_counts().sort_index()
print()
print('Rows per year:')
print(rows_per_year)

# (c) Find mean, min, and max wage for each year.
wage_summary = (
    Wage.groupby('year')['wage']
    .agg(mean_wage = 'mean', min_wage = 'min', max_wage = 'max')
    .round(2)
) # I like this style because it's closer to R xD
print()
print('Mean, min, and max wage by year:')
wage_summary
```

1. Answer: The years on record were 2003, 2004, 2005, 2006, 2007, 2008, and 2009; the number of rows per year were, in order, 513, 485, 447, 392, 386, 388, and 389; and the wage summary are shown directly above. The result does make sense overall: each year has (very) roughly similiar number of observations (from like 400 ish to 500 ish), and the wage distribution (at least with just the mean, min, and max) roughly tracks. You'd expect the minimum wage to be around $20-40k (minimum wage jobs) the max to be around $300k (although, in reality, if you include the CEOs like Elon Musk, this would probably be much higher), and the mean (just like in any wage data set) gets pulled higher because of the outliers towards the higher end (which is often why we look at median wage salary instead of mean).

2. Question: Simple Linear Regression on ISLP (Hitters): (a) Build a linear regression model for each of the following variables to predict salary: AtBat, Hits, HmRun, Runs, RBI, Walks, Years (b) Which of these performs the best in terms of evidence of being linearly associated with salary?

```python
# Load Hitters data.
Hitters = load_data('Hitters')

# Keep requested predictors and response, then remove missing Salary rows.
pred_cols = ['AtBat', 'Hits', 'HmRun', 'Runs', 'RBI', 'Walks', 'Years']
hitters_lr = Hitters[pred_cols + ['Salary']].dropna()

# Fit one linear model per predictor.
rows = []
for col in pred_cols:
    X = hitters_lr[[col]]
    y = hitters_lr[['Salary']]
    model = LinearRegression().fit(X, y)
    y_hat = model.predict(X)
    rows.append({
        'predictor': col,
        'intercept': float(model.intercept_.ravel()[0]),
        'slope': float(model.coef_.ravel()[0]),
        'mse': float(np.mean((y.values - y_hat) ** 2)),
        'r2': float(model.score(X, y))
    })

# Compare simple linear models.
hitters_results = pd.DataFrame(rows).sort_values(by = 'r2', ascending = False)
hitters_results
```

```python
# Show best predictor and plot each fitted line.
best_hitter = hitters_results.iloc[0]
print(
    f"Best predictor by R^2 (and no so coincidentally MSE) is {best_hitter['predictor']} "
    f"(R^2 = {best_hitter['r2']:.4f}, MSE = {best_hitter['mse']:.2f})."
)
n         = len(pred_cols)
ncols     = 3
nrows     = math.ceil(n / ncols)
fig, axes = plt.subplots(nrows = nrows, ncols = ncols, figsize   = (12, 3.5 * nrows))
axes      = axes.ravel()
for i, col in enumerate(pred_cols):
    X = hitters_lr[[col]]
    y = hitters_lr[['Salary']]
    model = LinearRegression().fit(X, y)
    y_hat = model.predict(X)
    axes[i].scatter(X[col], y['Salary'], alpha = 0.6)
    axes[i].plot(X[col], y_hat, color = 'red')
    axes[i].set_title(f'Salary vs {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Salary')
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])
plt.tight_layout()
plt.show()
```

2. Answer: Based on the table in the chunk above the above chunk and the plots shown above, the best-performing predictor is `RBI` with the largest $R^2$ (and not so coincidentally) the smallest MSE among these seven simple models. That variable has the strongest evidence of a linear association with `Salary` in this restricted predictor set. From our statistics class, the way we'd interpret is that about 20% of the variation in the salary is explained by RBI; we are allowed to compare $R^2$ and not necessary the degrees-of-freedom-adjusted ones because they all have the same number of degrees of freedom. From the $R = 0.4495$ value alone, we can also see that there is some moderate positive linear relationship between `RBI` and `Salary` (based on my experience in sports analytics, we'd consider this low of a correlation coefficient as really significant because there's so much variability and random noise in sports).

3. Question: Multiple Linear Regression ISLP (Credit): (a) Build a multiple linear regression model to predict the Rating. (b) Next build the single variable models and compare the difference. (c) How much better or worse in the multiple linear regression model at predicting the Rating versus the best single variable model?

```python
# Load Credit data.
Credit = load_data('Credit')

# (a) Multiple linear regression model for Rating.
y_credit = Credit['Rating']
X_credit_multi = pd.get_dummies(Credit.drop(columns = ['Rating']), drop_first = True)
credit_multi_model = LinearRegression().fit(X_credit_multi, y_credit)
credit_multi_hat = credit_multi_model.predict(X_credit_multi)
credit_multi_r2 = float(credit_multi_model.score(X_credit_multi, y_credit))
credit_multi_mse = float(np.mean((y_credit.values - credit_multi_hat) ** 2))
print(f'Multiple model R^2: {credit_multi_r2:.4f}')
print(f'Multiple model MSE: {credit_multi_mse:.4f}')

# (b) Single-variable models for comparison.
rows = []
for col in Credit.columns:
    if col == 'Rating':
        continue
    X_single = pd.get_dummies(Credit[[col]], drop_first = True)
    single_model = LinearRegression().fit(X_single, y_credit)
    y_hat = single_model.predict(X_single)
    rows.append({
        'predictor': col,
        'mse': float(np.mean((y_credit.values - y_hat) ** 2)),
        'r2': float(single_model.score(X_single, y_credit))
    })
credit_single_results = pd.DataFrame(rows).sort_values(by = 'r2', ascending = False)
print()
print('Single-variable model results (best at top):')
credit_single_results
```

```python
# (c) Compare multiple model to the best single-variable model.
best_credit_single = credit_single_results.iloc[0]
r2_gain = credit_multi_r2 - float(best_credit_single['r2'])
mse_drop = float(best_credit_single['mse']) - credit_multi_mse
print(
    f"Best single-variable model: {best_credit_single['predictor']} "
    f"(R^2 = {best_credit_single['r2']:.4f}, MSE = {best_credit_single['mse']:.4f})"
)
print(f'R^2 improvement (multiple - best single): {r2_gain:.4f}')
print(f'MSE reduction (best single - multiple): {mse_drop:.4f}')
```

3. Answer: While the multiple linear regression does better in terms of both $R^2$ and $MSE$, it's not by a considerable margin that I'd trade interpretability for. Just by using the limit, we capture 99.38% of the variation in the rating, which is way more than enough. I'm also pretty sure that if we were to use the degrees-of-freedom adjusted $R^2$ values, it'd show that the simple linear model with just the `Limit` is actually better than the full multiple linear model. Since we did not perform any train-test splitting, I'm worried that both of these models are overfit (it's almost too good) and/or that we don't actually have this information when predicting the rating in reality (i.e., leakage, although I don't know too much about the credit card world, so I wouldn't be able to comment on what's really happening here to create such high $R^2$ values).

4. Question: Multiple Linear Regression ISLP (College): (a) Build a model that can predict the sum of the Outstate, Room.Board, and Books columns (excluding the Expense variable). Here we are aggregating the separate costs for each school into one value and they are trying to predict it with the other variables. (b) check the variables to see if there are any potential interactions and then revise your model to have the interaction terms. (c) keep the best model of your results and explain if it is a good performing model or not.

```python
# Load college data.
College = load_data('College')

# Build response variable as requested.
College['TotalCost'] = College['Outstate'] + College['Room.Board'] + College['Books']

# (a) Build baseline predictors and exclude any of the expense colunns.
drop_cols  = ['Outstate', 'Room.Board', 'Books', 'TotalCost', 'Expend']
drop_cols  = [c for c in drop_cols if c in College.columns]
X_base_raw = College.drop(columns = drop_cols)
y_college  = College['TotalCost']
X_base     = pd.get_dummies(X_base_raw, drop_first = True)
base_model = LinearRegression().fit(X_base, y_college)
base_hat   = base_model.predict(X_base)
base_r2    = float(base_model.score(X_base, y_college))
base_mse   = float(np.mean((y_college.values - base_hat) ** 2))

# (b) Add interaction terms among selected numeric variables.
inter_cols  = [c for c in ['Accept', 'Enroll', 'S.F.Ratio', 'PhD', 'perc.alumni']]
X_inter_raw = X_base_raw.copy()
for i in range(len(inter_cols)):
    for j in range(i + 1, len(inter_cols)):
        c1 = inter_cols[i]
        c2 = inter_cols[j]
        X_inter_raw[f'{c1}_x_{c2}'] = X_inter_raw[c1] * X_inter_raw[c2]
X_inter = pd.get_dummies(X_inter_raw, drop_first = True)
inter_model = LinearRegression().fit(X_inter, y_college)
inter_hat   = inter_model.predict(X_inter)
inter_r2    = float(inter_model.score(X_inter, y_college))
inter_mse   = float(np.mean((y_college.values - inter_hat) ** 2))

# (c) Keep the best model.
college_compare = pd.DataFrame([
    {'model': 'Base model', 'r2': base_r2, 'mse': base_mse},
    {'model': 'Interaction model', 'r2': inter_r2, 'mse': inter_mse}
]).set_index('model')
college_compare
```

```python
# Summarize the best model.
best_model_name = college_compare['mse'].idxmin()
print('Best model by MSE:', best_model_name)
print(college_compare)
```

4. Answer: I created `TotalCost = Outstate + Room.Board + Books`, fit a baseline multiple linear model with all the other variables except `Expend`, then added some intuitive interactions like Accept * Enroll and S.F.Ratio * Enroll. The model with the lower MSE was the interaction model (not by any magnitude of difference, but considerable amount). Since this model has a fairly high $R^2=0.724562$ as well, it is performing well for this target (although, not to say the base model without the interactions is not much worse).
